# **Theoretical Questions:**

**Q1.** What is unsupervised learning in the context of machine learning?  
**Ans:**Unsupervised learning is a type of machine learning where the model is given data without labeled outputs and must find patterns, relationships, or structures in the data on its own. Unlike supervised learning, which relies on labeled training data to make predictions, unsupervised learning works with raw, unstructured data.  
Key Aspects of Unsupervised Learning:  
No explicit labels: The algorithm is not given predefined categories or labels.

Pattern discovery: It uncovers hidden structures or clusters within the data.

Self-organization: The model organizes data into meaningful representations

**Q2.** How does K-Means clustering algorithm work ?  
**Ans:** K-Means clustering is an unsupervised machine learning algorithm used for partitioning data into K distinct clusters.  
Steps:  
Initialization: Choose the number of clusters, K, and randomly initialize K centroids .  
Assignment: Assign each data point to the nearest centroid based on the Euclidean distance.  
Update: Compute the new centroids by taking the mean of all points in each cluster.  
Repeat: Continue steps 2 and 3 until the centroids stop changing .

How It Works:  
K-Means tries to minimize the intra-cluster variance, meaning that points within the same cluster should be as close as possible to each other.
It converges when the centroids stabilize and do not change across iterations.

Q3. Explain the concept of a dendrogram in hierarchical clustering?  
**Ans:**A dendrogram is a tree-like diagram that visualizes the arrangement of clusters in hierarchical clustering. It shows how data points (or objects) are grouped together in a nested manner, from individual elements to broader clusters.  
Key Aspects of a Dendrogram:  
Nodes and Branches: Each leaf node represents a single data point, and as you move up the hierarchy, nodes merge into clusters.

Height of Links: The vertical position of a merge indicates the similarity or distance between clusters—the lower the merge, the more similar the clusters.

Cutting the Dendrogram: By choosing a specific height (threshold), you can define the number of final clusters.

Q4. What is the main difference between K-Means and Hierarchical Clustering?  
**Ans:**   
**K-Means Clustering:**
It partitions data into a predefined number of clusters (K).
Uses an iterative optimization process to minimize the variance within each cluster.Clusters are formed based on centroids, which are recalculated as the algorithm runs.It is fast and scalable, making it ideal for large datasets.
Requires users to specify the number of clusters in advance.

**Hierarchical Clustering:**
It builds a hierarchy of clusters in a bottom-up (agglomerative) or top-down (divisive) manner.
Does not require specifying the number of clusters beforehand.
Creates a dendrogram, allowing users to visually select the number of clusters after analysis. Is more computationally expensive compared to K-Means, especially for large datasets.
Can capture nested and complex relationships better than K-Means.

**Q5.**What are the advantages of DBSCAN over K-Means?  
**Ans:** DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers several advantages over K-Means :

1) No Need to Predefine the Number of Clusters – Unlike K-Means, which requires specifying the number of clusters beforehand, DBSCAN automatically determines the number of clusters based on data density.

2) Detects Arbitrarily Shaped Clusters – K-Means tends to form spherical clusters, but DBSCAN can detect clusters of any shape, making it more flexible for complex datasets.

3) Handles Noise Well – DBSCAN marks outliers as noise instead of forcing them into a cluster, leading to better handling of noisy data compared to K-Means.


**Q6.** When would you use Silhouette Score in clustering?

Cluster Quality: Are data points well-grouped within their assigned clusters?

Separation Between Clusters: Is there a clear distinction between different clusters?

Choosing the Optimal Number of Clusters: When using algorithms like K-Means, Silhouette Score can help determine the best number of clusters by comparing different values of k.

**Q7.** What are the limitations of Hierarchical Clustering ?  
**Ans:** Hierarchical clustering is a powerful technique, but it comes with several limitations:

Scalability Issues: It has a high computational complexity, especially with large datasets. The time complexity is typically O(n²) or worse, making it inefficient for big data applications.

Fixed Structure: Once a hierarchy is built, it cannot be altered. This rigidity means that errors in early merging or splitting cannot be corrected.

Difficulty in Choosing the Number of Clusters: Unlike k-means, hierarchical clustering does not have a clear method to determine the optimal number of clusters, requiring manual intervention or external validation metrics.

**Q8.**Why is feature scaling important in clustering algorithms like K-Means?  
**Ans:**Feature scaling is crucial in clustering algorithms like K-Means because these algorithms use distance-based calculations to determine cluster assignments.

Equal Contribution: In raw data, features can have vastly different ranges. For example, if one feature represents income (in thousands) and another represents age (in years), the income values could dominate the clustering process. Scaling ensures all features contribute equally.

Improved Convergence: K-Means relies on minimizing the sum of squared distances between points and their assigned cluster centers. Without scaling, the algorithm may take longer to converge or even get stuck in suboptimal solutions.

Better Cluster Formation: When features are on different scales, clusters can be skewed or elongated along the dominant feature's axis. Scaling prevents this distortion and allows for more natural cluster formations.

**Q9.** How does DBSCAN identify noise points?  
**Ans:** DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points based on density criteria.

Core Points: DBSCAN considers a point as a core point if there are at least MinPts (a specified threshold) neighbors within a given radius (ε).

Border Points: If a point is within ε of a core point but does not itself have enough neighbors to be a core point, it is classified as a border point.

Noise Points: Any point that is neither a core point nor a border point is labeled as noise. These points do not belong to any cluster and are typically outliers or sparsely distributed.

**Q11.** What is the elbow method in K-Means clustering?  
**Ans:** The elbow method is a technique used to determine the optimal number of clusters (k) in K-Means clustering. It helps to identify the point where adding more clusters no longer provides significant improvement.

**Q12.**Describe the concept of "density" in DBSCAN .  
**Ans:** In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density is the core concept that determines how clusters are formed. It refers to the concentration of data points in a given region.

DBSCAN groups points based on how closely they are packed together, rather than relying on predefined cluster shapes. The algorithm defines density in terms of:

ε-neighborhood (epsilon radius): A circle around each point with radius ε.

MinPts (minimum points): The minimum number of points required within the ε-neighborhood for it to be considered dense.

**Q13.** Can hierarchical clustering be used on categorical data?  
**Ans:** Yes! Hierarchical clustering can be applied to categorical data, though it requires a suitable distance metric. Unlike numerical data, categorical data doesn't have a natural ordering, so standard Euclidean distance isn’t useful. Instead, here are some common approaches:

Hamming Distance: Counts the number of mismatched attributes between data points.

Jaccard Similarity: Measures similarity based on shared attributes, useful for binary data.

Gower Distance: A mixed-type distance measure that works well when categorical and numerical features are combined.

**Q14.**What does a negative Silhouette Score indicate?   
**Ans:** A negative Silhouette Score suggests that the distance between the data point and its assigned cluster is greater than the distance to another cluster. This indicates poor clustering, as the point might belong to a different group instead.

**Q15.**Explain the term "linkage criteria" in hierarchical clustering .  
**Ans:** In hierarchical clustering, linkage criteria determine how the distance between clusters is calculated when merging them. These criteria guide the clustering process, affecting the structure and formation of the resulting hierarchy.

There are several types of linkage criteria:  
Single Linkage: The distance between two clusters is defined by the closest pair of points between them.  
Complete Linkage: The distance is based on the farthest pair of points between the clusters.  
Average Linkage: The mean distance between all points across the two clusters is used.  
Centroid Linkage: The distance is determined by the centroids (mean position) of the clusters.

**Q16.**Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?   
**Ans:** K-Means clustering can struggle with datasets that have varying cluster sizes or densities because of its fundamental assumptions and mechanics. Here’s why:

Sensitivity to Cluster Sizes: K-Means assumes that all clusters are roughly equal in size, because it minimizes within-cluster variance. If clusters are significantly different in size, the algorithm may force points from a smaller cluster into a larger one, distorting results.

Sensitivity to Cluster Densities: K-Means relies on distance-based assignments, meaning denser clusters (where points are tightly packed) can be overshadowed by more dispersed clusters. As a result, it might incorrectly assign points to a nearby larger, sparser cluster instead of keeping them in a smaller, denser one.

Centroid-Based Approach: The algorithm represents each cluster by its centroid, which works well for circular or convex-shaped clusters. If clusters have irregular shapes or if densities vary, K-Means might place centroids in a way that doesn't accurately reflect true groupings.

**Q17.** What are the core parameters in DBSCAN, and how do they influence clustering?  
**Ans:** DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has two key parameters:

Epsilon (ε) – This defines the radius within which points are considered neighbors. If a point has enough nearby points within this radius, it's classified as a core point. Smaller values lead to more compact clusters, while larger values allow broader clusters but risk merging different groups.

MinPts – The minimum number of points required within the ε radius for a point to be considered a core point. Higher values lead to fewer, denser clusters, while lower values create more fragmented clusters and may classify more points as noise.

How They Influence Clustering:  
Epsilon too small → Many points may be classified as noise, leading to incomplete clusters.

Epsilon too large → Different clusters may merge, reducing the ability to separate distinct groups.

MinPts too small → Outliers may incorrectly be included in clusters.

MinPts too large → Many points may be labeled as noise, limiting cluster formation.

**Q18.** How does K-Means++ improve upon standard K-Means initialization?   
**Ans:** K-Means++ improves standard K-Means initialization by strategically selecting initial cluster centers, which helps prevent poor clustering results caused by bad starting positions.

Key Improvements:
Smarter Initialization – Instead of randomly choosing initial centroids, K-Means++ picks centers in a way that maximizes spacing between them. The first center is selected randomly, but subsequent centers are chosen with probability proportional to the squared distance from existing centers. This ensures well-distributed initial points.

Faster Convergence – Standard K-Means often gets stuck in local optima due to unlucky centroid placement. K-Means++ gives centroids a better starting position, leading to fewer iterations to reach a good solution.

Higher Stability – Random initialization in K-Means can produce vastly different results depending on the starting points. K-Means++ reduces this randomness, leading to more consistent and reproducible clusters.

**Q19.**What is agglomerative clustering?  
**Ans:**Agglomerative clustering is a type of hierarchical clustering that builds clusters step by step from the bottom up. It starts with each data point as its own cluster and then iteratively merges the most similar clusters until a stopping condition is met—often when all points are grouped into a single cluster or a predefined number of clusters is reached.

**Q20.**What makes Silhouette Score a better metric than just inertia for model evaluation?  
**Ans:** The Silhouette Score offers a more nuanced view of clustering performance compared to inertia. Here's why it's often considered a better metric:  
Relative vs. Absolute Evaluation:

Inertia measures the compactness of clusters by summing the squared distances of points to their assigned cluster centroids. While this can indicate tight clusters, it doesn't assess whether those clusters are actually well-separated.

Silhouette Score , on the other hand, considers both intra-cluster cohesion (how close points are within the same cluster) and inter-cluster separation (how distinct clusters are from each other). This dual consideration makes it more informative.

Scale Dependency:  
Inertia is highly dependent on the number of clusters and dataset scale. It generally decreases as the number of clusters increases, making it harder to determine the optimal number of clusters.

Silhouette Score is normalized between -1 and 1, making it easier to compare across different clustering solutions and dataset scales.

```python
#1.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, random_state=42, cluster_std=1.5)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('K-Means Clustering on Synthetic Data')
plt.show()
```

```python
#9.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='yellow')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k Values")
plt.xticks(k_values)
plt.show()
```

```python
#11. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist

X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.8, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

def plot_decision_boundaries(X, kmeans):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
    Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap='rainbow')
    plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, edgecolors='k', cmap='rainbow')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='X', c='black', edgecolors='white')
    plt.title("K-Means Clustering with Decision Boundaries")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

plot_decision_boundaries(X, kmeans)
```

```python
#20.  Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot.

import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np

# Load the Digits dataset
digits = datasets.load_digits()
data = digits.data

# Apply PCA to reduce to 3 components
pca = PCA(n_components=3)
reduced_data = pca.fit_transform(data)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
clusters = kmeans.fit_predict(reduced_data)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(reduced_data[:, 0], reduced_data[:, 1], reduced_data[:, 2], c=clusters, cmap='tab10')

ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("PCA-reduced Digits dataset with KMeans clustering")
plt.colorbar(scatter)
plt.show()
```

```python
#21.Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the clustering

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data with 5 centers
X, y = make_blobs(n_samples=500, centers=5, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
y_pred = kmeans.fit_predict(X)

# Calculate silhouette score
score = silhouette_score(X, y_pred)
print(f'Silhouette Score: {score:.2f}')

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='red', marker='X', label='Centers')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```

```python
#22.Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

# Load the dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Reduce dimensionality using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply Agglomerative Clustering
cluster = AgglomerativeClustering(n_clusters=2)
labels = cluster.fit_predict(X_pca)

# Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Agglomerative Clustering on Breast Cancer Dataset (2D PCA Projection)")
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
#23.Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Generate noisy circular data
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.05)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plot results side-by-side
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=20)
ax[0].set_title("KMeans Clustering")

ax[1].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', s=20)
ax[1].set_title("DBSCAN Clustering")

plt.show()
```

```python
#24.  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_samples

# Load Iris dataset
iris = load_iris()
X = iris.data

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

silhouette_vals = silhouette_samples(X, labels)

plt.figure(figsize=(8, 5))
y_lower = 10
for i in range(n_clusters):
    cluster_silhouette_vals = silhouette_vals[labels == i]
    cluster_silhouette_vals.sort()
    y_upper = y_lower + len(cluster_silhouette_vals)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_silhouette_vals)
    y_lower = y_upper + 10

plt.xlabel('Silhouette Coefficient')
plt.ylabel('Samples')
plt.title('Silhouette Coefficient for Each Sample')
plt.show()
```

```python
#26. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load Wine dataset
wine = load_wine()
X = wine.data[:, :4]  # Select first 4 features
y = wine.target  # Actual labels (not used in clustering)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
wine['cluster'] = kmeans.fit_predict(X_scaled)

# Create a DataFrame for visualization
df = pd.DataFrame(X, columns=wine.feature_names[:4])
df['cluster'] = wine['cluster']

# Visualize using Seaborn pairplot
sns.pairplot(df, hue='cluster', palette='Set1', diag_kind='kde')
plt.suptitle("KMeans Clustering of Wine Dataset (First 4 Features)")
plt.show()
```

```python
#27.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate synthetic data with noise
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X)

# Count clusters and noise points
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Excluding noise label (-1)
num_noise_points = list(labels).count(-1)

print(f"Number of clusters: {num_clusters}")
print(f"Number of noise points: {num_noise_points}")

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30)
plt.title("DBSCAN Clustering with Noisy Blobs")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
#28.Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = datasets.load_digits()
data = digits.data

tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(data)

cluster = AgglomerativeClustering(n_clusters=10)
labels = cluster.fit_predict(data_tsne)

plt.figure(figsize=(8, 6))
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.colorbar(label='Cluster Label')
plt.title('Agglomerative Clustering on t-SNE Reduced Digits Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
